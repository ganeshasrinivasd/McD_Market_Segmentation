Reading Data

```python
url="https://homepage.boku.ac.at/leisch/MSA/datasets/mcdonalds.csv"
df = pd.read_csv(url)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

for col in df.columns:
  df[col] = encoder.fit_transform(df[col])
```

```python
from yellowbrick.features.pca import PCADecomposition

visualizer = PCADecomposition(proj_features=True, alpha=0.3)
visualizer.fit_transform(df)
visualizer.show()
```

```python
categorical = []
continuous = []
for i in df:
    if(df[i].dtypes == "object"):
        categorical.append(i)
    else:
        continuous.append(i)
```

```python
y=df[["yummy"]]
x=df.drop(labels=["yummy"],axis=1)
```

```python
from warnings import filterwarnings
filterwarnings("ignore")
for i in x.columns:
    if(x[i].dtypes=="object"):
        sb.countplot(x[i],hue=y.yummy)
        plt.show()
    else:
        sb.boxplot(y.yummy,x[i])
        plt.show()
```
